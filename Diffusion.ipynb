```python
import numpy as np
import random
import math
import matplotlib.pyplot as plt
```

## 2D diffusion

```python
numsteps = 100
rx =[0]
ry = [0]
for i in range(numsteps):
    rxi = random.uniform(-1.0,1.0)
    ryi = random.uniform(-1.0,1.0)
    norm = math.sqrt(rxi**2 + ryi**2)
    rxi = rxi/norm
    ryi = ryi/norm
    rx.append(rx[i]+rxi)
    ry.append(ry[i]+ryi)

plt.figure(figsize=(10, 10))
plt.plot(rx, ry, linestyle='-', marker='.')
plt.title('2D Random Walk')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

## 3D diffusion

```python
from mpl_toolkits.mplot3d import Axes3D
numsteps = 1000
rx = [0]
ry = [0]
rz = [0]
```

```python
for i in range(numsteps):
    u = random.uniform(0,1)
    v = random.uniform(0,1)
    theta = 2*np.pi*u
    phi = np.arccos(2*v - 1)
    rx.append(rx[i]+np.cos(phi)*np.sin(theta))
    ry.append(ry[i]+ np.sin(phi)*np.sin(theta))
    rz.append(rz[i]+np.cos(phi))
    
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot(rx, ry, rz, linestyle='-', marker='.')
ax.set_title('3D Random Walk')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

    
```
